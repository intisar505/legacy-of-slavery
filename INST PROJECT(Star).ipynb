```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('Legacy of Slavery.csv', low_memory=False)
df['County'].value_counts().head()
pd.value_counts(df['County']).head().plot.bar()
plt.title("Highest Populated Counties by Slaves")
plt.xlabel("Counties")
plt.ylabel("Number of Slaves in Population")
```

```python
df["reward_amount"].value_counts().head()
pd.value_counts(df["reward_amount"]).head().plot.bar()

plt.title("Average reward amount")
plt.xlabel("reward amount")
plt.ylabel("Number of Slaves in Population")
```

```python
df['destination_justification'].value_counts(10-15)
pd.value_counts(df['destination_justification']).head().plot.bar()
plt.title("runaway justifications")
plt.xlabel("justifications")
plt.ylabel("Number of Slaves in Population")
#THIS GRAPH MAKES NO SENSE BECAUSE THEIR REASONS FOR RUNNIG AWAY ARE ALL DIFFERENT AND VERY SPECIFIC
```

```python
df['fugative_age'].value_counts().head()
pd.value_counts(df['fugative_age']).head().plot.bar()
plt.title("average age of slaves")
plt.xlabel("age")
plt.ylabel("Number of Slaves in Population")
```
